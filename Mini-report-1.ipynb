## Bisection Method

```python
def bisection_step(f, bounds):
    '''defines bisection_step with arguments f and bounds'''
    lower, upper = bounds     
    middle = (lower + upper)/2
    
    if f(lower)*f(middle)<=0:                     #if loop printing lower and middle values if f(l)*f(m)<0 otherwise printing the middle and upper values
        return (lower,middle)
    else:
        return(middle,upper) 
```

```python
def bisection(f,Range):
    '''defines bisection with arguments f and Range'''
    l, u = Range
    n=1
    while u-l>=(10**-8):                          #while loop printing bisection_step functions while the upper value - the lower value is greater than 10**-8
        l, u = bisection_step(f, (l, u)) 
        n=n+1
        print("{:15.12f}  {:15.12f}  {:15.12f}".format(l, u, u-l),'(',n,')')
```

```python
print("{:^15}  {:^15}  {:^15}".format("lower", "upper", "difference"))
print('Bisection Method for a(x)')               #finds root for a(x) using the bisection function above in the range 0 to 5
bisection(a,Range=(0,5))

print('Bisection Method for b(x) (first root)')  #finds first root for b(x) using the bisection function above in the range -0.2 to -1
bisection(b,Range=(-1,-0.2))

print('Bisection Method for b(x) (second root)') #finds second root for b(x) using the bisection function above in the range 0.7 to 1.4
bisection(b,Range=(0.7,1.4))

print('Bisection Method for c(x)')               #finds root for c(x) using the bisection function above in the range 0 to 1.5
bisection(c, Range=(0,1.5))

print('Bisection Method for d(x) (first root)')  #finds first root for d(x) using the bisection function above in the range 0 to 1.3
bisection(d, Range=(0,1.3))

print('Bisection Method for d(x) (second root)') #finds second root for d(x) using the bisection function above in the range 0 to 1.3
bisection(d, Range=(1.7,2.4))

print('Bisection Method for e(x) (first root)')  #finds first root for e(x) using the bisection function above in the range 0 to 1.5
bisection(e,Range=(0,1.5))

print('Bisection Method for e(x) (second root)') #finds second root for e(x) using the bisection function above in the range 1.5 to 3.5
bisection(e,Range=(1.5,3.5))

print('Bisection Method for f(x)')               #finds root for f(x) using the bisection function above in the range -5 to 0
bisection(f,Range=(-5,0))

print('Bisection Method for g(x)')               #finds root for f(x) using the bisection function above in the range 0 to 2
bisection(g,Range=(0,2))
```

## Regula Falsi

```python
def regula_falsi_step(f, bounds):
    '''defines regula_falsi_step with arguments f and Bounds'''
    lower, upper = bounds
    #middle = (lower + upper)/2
    middle=((lower*f(upper))-(upper*f(lower)))/(f(upper)-f(lower))

    if f(lower)*f(middle)<0:     #if loop printing lower and middle values if f(l)*f(m)<0 otherwise printing the middle and upper values
        return (lower,middle)
    else:
        return(middle,upper)
        
```

```python
def regula_falsi(f,a,b):
    '''defines regula_falsi with arguments f and Range'''
    l, u = a,b
    n=0
    while abs(f(l))>(10**-8) and abs(f(u))>(10**-8): #while loop printing regula_falsi_step functions while f(l)>10**-8 and f(u)>10**-8
        l, u = regula_falsi_step(f, (l, u)) 
        n=n+1
        if n>1000:
            break
        else:
            print("{:15.12f} {:15.12f}".format(l, u, u-l),'(',n,')') 
        #print("{:15.12f} {:15.12f}".format(l, u, u-l),'(',n,')') 

```

## Newton Raphson Method

```python
def newton_step(f, fp, x0):
    '''defines newton_step with arguments f, fp and x0'''
    return (x0-((f(x0))/fp(x0)))
```

```python
def a(x):
    '''defines a(x) to be x**2 - 4'''
    return x**2 - 4
def ap(x):
    '''defines ap(x) to be 2*x as the derivative of a(x)'''
    return 2*x

def b(x):
    '''defines b(x) to be ((x**8)-(2*x)-1)'''
    return ((x**8)-(2*x)-1)
def bp(x):
    '''defines bp(x) to be (8*(x**7))-2 as the derivative of b(x)'''
    return (8*(x**7))-2

def c(x):
    '''defines c(x) to be ((x-1)/(x-2))'''
    return ((x-1)/(x-2))
def cp(x):
    '''defines cp(x) to be (-(1/((x-2)**2))) as the derivative of c(x)'''
    return (-(1/((x-2)**2)))

def d(x):
    '''defines d(x) to be ((x**3)+(47*x**2)-(148*x)+90)'''
    return ((x**3)+(47*x**2)-(148*x)+90)
def dp(x):
    '''defines dp(x) to be (3*(x**2))+(94*x)-148 as the derivative of d(x)'''
    return (3*(x**2))+(94*x)-148

def e(x):
    '''defines e(x) to be ((x**4)-(8*x**3)+(22*x**2)-(24*x)+9)'''
    return ((x**4)-(8*x**3)+(22*x**2)-(24*x)+9)
def ep(x):
    '''defines ep(x) to be (4*(x**3))-(24*(x**2))+(44*x)-24 as the derivative of e(x)'''
    return (4*(x**3))-(24*(x**2))+(44*x)-24

def f(x):
    '''defines f(x) to be ((x**3)-(2*x)+2)'''
    return ((x**3)-(2*x)+2)
def fp(x):
    '''defines fp(x) to be (3*(x**2))-2 as the derivative of f(x)'''
    return (3*(x**2))-2

def g(x):
    '''defines f(x) to be cos(x)'''
    return cos(x)
def gp(x):
    '''defines fp(x) to be -sin(x) as the derivative of f(x)'''
    return -sin(x)


x0=1
x1=0
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range 1 to 0
    n=n+1
    x1=x0
    x0=newton_step(a,ap,x0)
print('a',x0,"(took",n, "iterations)")

x0=-1
x1=-0.2
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range -1 to -0.2
    n=n+1
    x1=x0
    x0=newton_step(b,bp,x0)
print('b1',x0,"(took",n, "iterations)")

x0=1
x1=0
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range 0.7 to 1.4
    n=n+1
    x1=x0
    x0=newton_step(b,bp,x0)
print('b2',x0,"(took",n, "iterations)")

x0=1
x1=0
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range 1 to 0
    n=n+1
    x1=x0
    x0=newton_step(c,cp,x0)
print('c',x0,"(took",n, "iterations)")

x0=1
x1=0
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range 1 to 0
    n=n+1
    x1=x0
    x0=newton_step(d,dp,x0)
print('d1',x0,"(took",n, "iterations)")

x0=1.7
x1=1
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range 1.7 to 1
    n=n+1
    x1=x0
    x0=newton_step(d,dp,x0)
print('d2',x0,"(took",n, "iterations)")

x0=0
x1=1.5
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range 0 to 1.5
    n=n+1
    x1=x0
    x0=newton_step(e,ep,x0)
print('e1',x0,"(took",n, "iterations)")

x0=2.5
x1=3.2
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range 1.5 to 3.5
    n=n+1
    x1=x0
    x0=newton_step(e,ep,x0)
print('e2',x0,"(took",n, "iterations)")

x0=-2
x1=0
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range -2 to 0
    n=n+1
    x1=x0
    x0=newton_step(f,fp,x0)
print('f',x0,"(took",n, "iterations)")

x0=1.1
x1=1.6
n=0
while abs(x0-x1)>(10**-8):                        #while loop printing newton_step functions while x0-x1>10**-8 in the range 0 to 2
    n=n+1
    x1=x0
    x0=newton_step(g,gp,x0)
print('g',x0,"(took",n, "iterations)")



```

```python
def newton_raphson(f,fp,x0):
    '''defines newton_raphson with arguments f, fp and x0'''
    x=x0
    x1=1
    while abs(x-x1)>10**-8:                        #while loop printing newton_step functions while x-x1>10**-8 in the range
        x1=x
        x=newton_step(f,fp,x)
    return x

print('Newtons Method for a(x)')
#print(newton_raphson(a,ap,0))
print(newton_raphson(a,ap,1))
print(newton_raphson(a,ap,0.14))
print(newton_raphson(a,ap,0.141))

print('Newtons Method for b(x)')
print(newton_raphson(b,bp,0))
print(newton_raphson(b,bp,1))
print(newton_raphson(b,bp,0.14))
print(newton_raphson(b,bp,0.141))

print('Newtons Method for c(x)')
#print(newton_raphson(c,cp,0))
print(newton_raphson(c,cp,1))
print(newton_raphson(c,cp,0.14))
print(newton_raphson(c,cp,0.141))

print('Newtons Method for d(x)')
print(newton_raphson(d,dp,0))
print(newton_raphson(d,dp,1))
print(newton_raphson(d,dp,0.14))
print(newton_raphson(d,dp,0.141))

print('Newtons Method for e(x)')
print(newton_raphson(e,ep,0))
print(newton_raphson(e,ep,1))
print(newton_raphson(e,ep,0.14))
print(newton_raphson(e,ep,0.141))

print('Newtons Method for f(x)')
#print(newton_raphson(f,fp,0))
print(newton_raphson(f,fp,1))
#print(newton_raphson(f,fp,0.14))
print(newton_raphson(f,fp,0.141))

print('Newtons Method for g(x)')
#print(newton_raphson(g,gp,0))
print(newton_raphson(g,gp,1))
print(newton_raphson(g,gp,0.14))
print(newton_raphson(g,gp,0.141))
```
